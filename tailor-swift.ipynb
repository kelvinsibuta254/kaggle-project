```python
import matplotlib.pyplot as plt
```

**Tailor Swift Popular Songs**

```python
df= pd.read_csv(
    "/kaggle/input/taylor-swift-spotify-data/spotify_taylorswift.csv",
    index_col= 0,
    parse_dates=["release_date"]

)
print("df shape:", df.shape)
print(df.info())
df.head()
```

```python
mask= df["popularity"]> 0
df[mask].sort_values("popularity", ascending=True).head()
```

Relationship between Popularity and Danceability

```python
plt.scatter(x= df[mask]["danceability"], y= df[mask]["popularity"])
plt.xlabel("Danceability")
plt.ylabel("Popularity")
plt.title("Tailor Swift Popularity vs Danceability");
```
